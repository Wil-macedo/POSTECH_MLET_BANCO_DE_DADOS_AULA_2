# O que é NoSQL <h6>

* O no SQl surgiu aproximadamente em 2009
* Surgiu para suprir as necessidades que os bancos SQL não conseguiam suprir:
    - Maior quantidade de dados
    - Mais recursos de armazenamento
    - Mais recursos para a leitura dos dados.



# Diferenção entre banco de dados relacionao e NÃO relacional:

# BD RELACIONAIS<H6>                                #BD NÃO RELACIONAIS<H6>

    - Padrão ANSI                                   - Não seguem um padrão
    - Linguagem SQL                                 - Tem sua própria linguagem de consulta
    - Armazenamento em tabelas                      - Tem sua própria estrutura de armazenamento
    

O NoSQL, ou Not Only SQL, é uma categoria de bancos de dados projetados para armazenar e recuperar informações de forma eficiente em formatos não relacionais, como documentos, chave-valor, colunas e grafos. Ao contrário dos bancos de dados relacionais, o NoSQL oferece flexibilidade na estrutura dos dados, permitindo que os documentos tenham diferentes campos e estruturas, sem a necessidade de um esquema rígido. Esse tipo de banco de dados é especialmente adequado para aplicações que exigem escalabilidade, velocidade, grandes volumes de dados e flexibilidade no esquema. Um exemplo popular de banco de dados NoSQL é o MongoDB, que utiliza o formato JSON para armazenar e organizar os dados. Ao considerar o uso de um banco de dados NoSQL, é importante entender as características, benefícios e boas práticas associados a ele para aproveitar ao máximo suas vantagens.

Geralmente armazenados no formato JSON(JavaScript Object Notation) ou XML (eXtensible Markup Language).
